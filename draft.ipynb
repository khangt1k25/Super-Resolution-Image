```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bn = [0, 1, 2]
color =['r', 'g', 'b', 'y']
```

```python
option =  "DLoss"
```

```python
plots = []
for i in bn:
    with open(f'./experiments/SRGAN/L1orL2/{option}_{i}.txt') as f:
        lines = f.read().splitlines()
        lines = [float(ele) for ele in lines]
        plots.append(lines)
        print(len(lines))
```

```python
plt.figure(figsize=(10, 5))
plt.title(f"{option} with the depth of net")
line1 = plt.plot(plots[0], color[0], label="0.5*l1+0.5*l2")
line2 = plt.plot(plots[1], color[1], label="l1")
line3 = plt.plot(plots[2], color[2], label="l2")
# lint4 = plt.plot(plots[3], color[3], label="9")
plt.xlabel("Epochs")
plt.ylabel(f"{option}")
plt.legend()
plt.savefig(f'./experiments/SRGAN/L1orL2/{option}.pdf')
#plt.show()
```
